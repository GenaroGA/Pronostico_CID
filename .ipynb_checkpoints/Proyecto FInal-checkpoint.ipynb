### Tratamiento de outliers

```r
attach(df_sem1)
```

```r
boxplot(df_sem1$devoluciones,main='Box Plot Devoluciones del CID Primavera 2019 (Grupos = 6)')
```

```r
grubbs.test(devoluciones)
grubbs.test(devoluciones,opposite = TRUE)
```

Tras realizar el test de Grubbs sobre los puntos extremos de la serie, se obtiene un $p-vale < 0.05$ para el valor máximo, comprobando que es un valor atípico. Para tratar este valor atípico se le realizará una transformación a la serie de tiempo, para determinar dicha transformación se calculará una $\lambda$.

```r
lambda <- BoxCox.lambda(devoluciones)
lambda
```

Como la $\lambda$ obtenida es cercana a 0 se realizara una transformación logarítmica a la serie.

```r
attach(df_sem1)
```

```r
df_sem1$dev_log <- log(devoluciones)
ts_dev <- xts(dev_log, order.by = fecha)
plot(ts_dev,main='Devoluciones del CID Primavera 2019',type='b')
```

```r
grubbs.test(dev_log)
grubbs.test(dev_log,opposite = TRUE)
```

Apesar de que en el grafico se apresia un suavizamiento de la serie, el test de Grubbs compruba que existe un valor atípico en el valor mínimo. Para mantener la simplicidad del modelo ya no se realizará una segunda transformación a la serie, en su lugar se tratará de forma local el outlier.

```r
boxplot(df_sem1$dev_log,main='Box Plot Devoluciones del CID Primavera 2019 (Grupos = 6)')
```

```r
valor_buscado = min(df_sem1$dev_log)
posicion <- which(df_sem1 == valor_buscado, arr.ind = TRUE)
posicion
```

```r
df_sem1 <- df_sem1[-1,]
```

```r
head(df_sem1)
```

Como el valor atípico se tratába del dato de mayór antigüedad se optó por eliminarlo.

```r
grubbs.test(df_sem1$dev_log)
grubbs.test(df_sem1$dev_log,opposite = TRUE)
```

Después de una transformación y tras haber tratado localmente un dato el Test de Grubbs da un $p-value > 0.05$ para todos los valores extremos, dando por finalizado el tratamiento de los outliers.

```r
attach(df_sem1)
```

## Metodologia de Box Jenkins

```r
ts_dev <- xts(dev_log, order.by = fecha)
plot(ts_dev,main='Devoluciones del CID Primavera 2019',type='b')
```

A simple vista la grafia de la serie no parece estacionaria, pues su media no se aprecia constante, así como su varianza.

```r
acf(dev_log,main="ACF Devoluciones del CID",lag.max=60)
```

La gráfica de la ACF respalda la afirmación anterior, pues esta decrece lentamente indicando no estacionalidad en la serie.

```r
spec<-spectrum(dev_log,main="Periodograma de las devoluciones en la primavera 2019",
               xlab="Frecuencia",ylab="Espectro")
threshold <- quantile(spec$spec, 0.95)

frecuencias_destacadas <- spec$freq[spec$spec >= threshold]
periodos_destacados <- 1/spec$freq[spec$spec >= threshold]

plot(spec, main = "Periodograma de la serie de devoluciones en la primavera 2019", xlab = "Frecuencia",
     ylab = "Densidad")
abline(h = threshold, col = "red", lty = 2)
points(frecuencias_destacadas, spec$spec[spec$spec >= threshold], col = "blue", pch = 19)
```

```r
periodos_destacados
```

### Tratamiento de la varianza

Se analizara la varianza de la serie con el fin de establizarla en caso de ser necesario para cumplir las condiciones de pronóstico de la metodología de Box Jenkins

```r
boxplot(df_sem1$dev_log~df_sem1$grupos,main='Box Plot Devoluciones del CID en la Primavera 2019 (Grupos = 6)')
```

```r
leveneTest(df_sem1$dev_log~df_sem1$grupos)
```

El Test de Levene da como resultado un $p-value > 0.05$ demostrando que la varianza es constante

### Tratamiento de la Tendencia

Ahora se analizara la tendencia para estabilizar la media.

```r
adf.test(df_sem1$dev_log,alternative = "stationary")
```

```r
dif1 <- diff(df_sem1$dev_log,1)
plot(dif1,main='Devoluciones del CID de la Primavera 2019',type='l')
```

```r
adf.test(dif1,alternative = "stationary")
```

### Proposición del Modelo

```r
acf(dif1,main="ACF Prestamos del CID",lag.max=10)
pacf(dif1,main="PACF Prestamos del CID",lag.max=10)
```

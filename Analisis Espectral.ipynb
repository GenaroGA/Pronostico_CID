## Análisis previo

Importación de datos

```r
library(readxl)
Prestamos_CID <- read_excel("Datos/Prestamos_Devoluciones_CID.xlsx",sheet = "Prestamos")
Devoluciones_CID <- read_excel("Datos/Prestamos_Devoluciones_CID.xlsx",sheet = "Devoluciones")
```

## Graficas

### Graficación de los datos

```r
plot(Prestamos_CID,type='l')
plot(Devoluciones_CID,type='l')
```

### ACFs

```r
acf(Prestamos_CID$cnt,plot ='true',lag.max=100000)
acf(Devoluciones_CID$cnt,plot ='true',lag.max=100000)
```

### PACFs

```r
pacf(Prestamos_CID$cnt,plot ='true',lag.max=100000)
pacf(Devoluciones_CID$cnt,plot ='true',lag.max=100000)
```

### Periodograma

```r
library('TSA')
periodograma_prestamos <-periodogram(Prestamos_CID$cnt,log='yes',plot='true')
plot(periodograma_prestamos)
```

```r
periodograma_devoluciones <-periodogram(Devoluciones_CID$cnt,log='yes',plot='true')
plot(periodograma_devoluciones)
```

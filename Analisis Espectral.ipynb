## Análisis previo

Importación de datos

```r
library(readxl)
Prestamos_CID <- read_excel("Datos/Prestamos_Devoluciones_CID.xlsx",sheet = "Prestamos2019")
Devoluciones_CID <- read_excel("Datos/Prestamos_Devoluciones_CID.xlsx",sheet = "Devoluciones2019")
```

## Graficas

### Graficación de los datos

```r
plot(Prestamos_CID,type='l',xlim=c(15,25))
plot(Devoluciones_CID,type='l',xlim=c(0,70))
```

### ACFs

```r
acfobj<-acf(Prestamos_CID$cnt,plot ='true',lag.max=24)
acf(Devoluciones_CID$cnt,plot ='true',lag.max=24)
```

### PACFs

```r
pacf(Prestamos_CID$cnt,plot ='true',lag.max=60)
pacf(Devoluciones_CID$cnt,plot ='true',lag.max=60)
```

### Periodograma

```r
#install.packages("descomponer")
library("descomponer")
gperiodograma(Prestamos_CID$cnt)
df_period_pres<-periodograma(Prestamos_CID$cnt)
head(df_period_pres,n=41)
```

```r
gperiodograma(Devoluciones_CID$cnt)
df_period_dev <- periodograma(Devoluciones_CID$cnt)
head(df_period_dev,n=40)
```
